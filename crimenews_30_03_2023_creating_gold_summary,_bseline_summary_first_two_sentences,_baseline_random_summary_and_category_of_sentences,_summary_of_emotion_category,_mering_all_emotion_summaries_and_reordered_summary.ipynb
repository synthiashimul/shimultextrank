```python
import pandas as pd
import numpy as np
import torch
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertModel, BertTokenizer
import nltk

# Load the BERT model and tokenizer
model = BertModel.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Define functions for preprocessing and tokenizing text
def preprocess_text(text):
    # Add any preprocessing steps here
    return text

def tokenize_into_sentences(text):
    sentences = nltk.sent_tokenize(text)
    return sentences

# Create a new column to store the generated summaries
df['woc_summary'] = np.nan

# Compute summary for each row of text column
for i, text in enumerate(df['text']):
    # Preprocess the text
    preprocessed_text = preprocess_text(text)

    # Tokenize the text into sentences
    sentences = tokenize_into_sentences(preprocessed_text)

    # Calculate the number of sentences to include in the summary
    default_ratio = 0.40
    num_sentences = max(round(default_ratio * len(sentences)), 1)

    # Compute sentence embeddings using BERT for each sentence
    sentence_embeddings = []
    for sentence in sentences:
        input_ids = tokenizer.encode(sentence, add_special_tokens=True)
        with torch.no_grad():
            last_hidden_states = model(torch.LongTensor([input_ids[:512]]))[0]
        if len(input_ids) > 512:
            for j in range(512, len(input_ids), 512):
                with torch.no_grad():
                    chunk_hidden_states = model(torch.LongTensor([input_ids[j:j+512]]))[0]
                last_hidden_states = torch.cat((last_hidden_states, chunk_hidden_states), dim=1)
        sentence_embeddings.append(last_hidden_states[:,0,:].numpy())

    # Compute cosine similarity between sentences
    similarity_matrix = cosine_similarity(np.concatenate(sentence_embeddings, axis=0))

    # Compute the  PageRank of each sentence
    n = similarity_matrix.shape[0]
    d = 0.85
    scores = np.ones(n) / n  # Reshape scores array to (n,)
    for _ in range(10):
        scores = (1 - d) + d * np.dot(similarity_matrix.T, scores)

     # Select the most important sentences based on their PageRank scores
    important_sentence_indices = np.argsort(scores)[-num_sentences:]
    important_sentences = [sentences[i] for i in important_sentence_indices]

    # Generate the summary
    summary = ' '.join(important_sentences)


    # Store the generated summary in the gen_summary column for the current row
    df.at[i, 'woc_summary'] = summary
```

```python
import random

def get_sentiment(sentences, language):
    emotions_list = []
    polarity_list = []
    emotion_class_list = []
    for sentence in sentences:
        emotions = {"EMOTION_"+label: 0 for label in EMOTIONSLABELS}
        polarity = 0
        for lemma in sentence['lemmas']:
            try:
                bsn[language].concept(lemma)
            except:
                continue
            moodtags = bsn[language].moodtags(lemma)
            polarity += float(bsn[language].polarity_value(lemma))
            for label in moodtags:
                emotions["EMOTION_"+EMOTIONSLABELS[EMOTIONS[language].index(label[1:])]] += 1
        # Find the emotion labels with the maximum value
        max_emotion_labels = [label for label in emotions if emotions[label] == max(emotions.values())]
        if len(max_emotion_labels) == 1:
            # No tie, use the emotion label with the maximum value
            max_emotion_label = max_emotion_labels[0]
        else:
            # Tie, randomly select one of the tied labels
            max_emotion_label = random.choice(max_emotion_labels)
        # Extract the emotion class label
        emotion_class = max_emotion_label.split("_")[1]
        emotions_list.append(emotions)
        polarity_list.append(polarity)
        emotion_class_list.append(emotion_class)
    return emotions_list, polarity_list, emotion_class_list
```

```python
# Define functions for preprocessing and tokenizing text
def preprocess_text(text):
    # Add any preprocessing steps here
    return text

def tokenize_into_sentences(text):
    sentences = nltk.sent_tokenize(text)
    return sentences
#pyramid score calculation
def pyramid_score(ref, hypo):
    # Tokenize the sentences into lists of words
    ref_tokens = ref.lower().split()
    hypo_tokens = hypo.lower().split()

    # Calculate unigram overlap
    ref_unigrams = Counter(ref_tokens)
    hypo_unigrams = Counter(hypo_tokens)
    num_unigrams = sum(hypo_unigrams.values())
    overlap_unigrams = sum(min(ref_unigrams[word], hypo_unigrams[word]) for word in hypo_unigrams)

    # Calculate bigram overlap
    ref_bigrams = Counter(zip(ref_tokens, ref_tokens[1:]))
    hypo_bigrams = Counter(zip(hypo_tokens, hypo_tokens[1:]))
    num_bigrams = sum(hypo_bigrams.values())
    overlap_bigrams = sum(min(ref_bigrams[word], hypo_bigrams[word]) for word in hypo_bigrams)

    # Calculate trigram overlap
    ref_trigrams = Counter(zip(ref_tokens, ref_tokens[1:], ref_tokens[2:]))
    hypo_trigrams = Counter(zip(hypo_tokens, hypo_tokens[1:], hypo_tokens[2:]))
    num_trigrams = sum(hypo_trigrams.values())
    overlap_trigrams = sum(min(ref_trigrams[word], hypo_trigrams[word]) for word in hypo_trigrams)

    # Calculate the Pyramid Score
    precision = (overlap_unigrams + overlap_bigrams + overlap_trigrams) / (num_unigrams + num_bigrams + num_trigrams)
    recall = (overlap_unigrams + overlap_bigrams + overlap_trigrams) / (sum(ref_unigrams.values()) + sum(ref_bigrams.values()) + sum(ref_trigrams.values()))
    if precision + recall > 0:
        pyramid_score = 2 * precision * recall / (precision + recall)
    else:
        pyramid_score = 0

    return pyramid_score

# Load the dataset
df = pd.read_csv('reorder_text.csv')

# Initialize Rouge
rouge = Rouge()

# Calculate Rouge-L F1 score, Bert F1 score, Cosine similarity, and Mover Score for each row and store the result in new columns
wrl_f1_scores = []
wbert_f1_scores = []
wcosim_scores = []
wmover_scores = []
wpyr_scores = []  # to store the Pyramid scores
for index, row in df.iterrows():
    final_summary = preprocess_text(row['wocord_summary'])
    g_summary = preprocess_text(row['gold_summary'])
    
    # Calculate wRouge-L F1 score
    scores_rouge = rouge.get_scores(final_summary, g_summary)[0]
    f1_rouge_l = scores_rouge['rouge-l']['f']
    wrl_f1_scores.append(f1_rouge_l)
    
    # Calculate wBert F1 score
    P, R, F1 = bert_score([final_summary], [g_summary], lang='en', model_type='bert-base-uncased', verbose=False)
    wbert_f1_scores.append(F1.item())
    
    # Calculate wCosine similarity
    sentences = [g_summary, final_summary]
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform(sentences).todense()
    similarity = cosine_similarity(np.asarray(vectors[0]), np.asarray(vectors[1]))[0][0]
    wcosim_scores.append(similarity)
    
    # Calculate wMover Score
    ref_tokens = nltk.word_tokenize(g_summary.lower())
    hypo_tokens = nltk.word_tokenize(final_summary.lower())
    mover_score = single_meteor_score(ref_tokens, hypo_tokens)
    wmover_scores.append(mover_score)

    # Calculate wPyramid Score
    pyr_score = pyramid_score(g_summary, final_summary)
    wpyr_scores.append(pyr_score)

df['wR-L-F1 score'] = wrl_f1_scores
df['wBERTF1'] = wbert_f1_scores
df['wcosim'] = wcosim_scores
df['wmovsco'] = wmover_scores
df['wpyrsco'] = wpyr_scores  # Add the Pyramid scores to the dataframe

# Save the modified dataframe
df.to_csv('reorder_text.csv', index=False)
```
